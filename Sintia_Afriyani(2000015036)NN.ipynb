**Model Neural Networks Dasar**

Tensor flow adalah sebuah library pada python untuk Deep learning. Pada bagian temen temen akan fokus pada syntaks bagaimana melakukan permodelan Neural Network Menguunakan library TensorFlow python. Pada bagian ini kita akan mulai pada kasus yang paling sederhana yaitu tabel logika "And" sebagai berikut:

 Perhatikan bahwa X1 dan X2 adalah input dan y adalah output yang diinginkan . maka ukuran input adalah 2 kolom an 4 baris sedangkan input adalah sebuah nilai antara 0 dan 1.

```python
 # import library 
 import tensorflow as tf
 from tensorflow import keras
 import numpy as np
```

```python
# Menampilkan versi Tensorflow
tf.__version__
```

Tiga tahap penting 
1. menyusun arsitektur model
2. mengkompilasi  model
3. training model
4. evaluasi model

```python
# Menyusun Model
model = keras.Sequential([                                                 #membentuk model squential
    tf.keras.layers.Dense(units=1,
                         input_shape=[2],
                         activation='relu'),                                #model layer pada NN
])
#compile model 
model.compile(loss='mean_square_error',                                     #menentukan fungsi loss
              optimizer='sgd')                                              #menentukan fungsi optimizer

model.summary()                                                             #menampilkan summary model
```

Berdasarkan model summary di atas maka kita sedang membuat model NN dengan satu titik layer. Kemudian hasil parameternya adalah tiga. Itu berarti terdapat  tiga parameter, apa aja???  W1 , W2 karwna inputnya adalah X1 dan X2 serta b yaitu bias. Berikut adalah visualisasi  NN modelnya

WhatsApp Image 2021-12-08 at 23.44.37.jpeg

```python
#Mendefinisikan Data Input(x) & Output(Y)
x = np.array([[1,1],[1,0],[0,1],[0,0]],dtype="int")
y = np.array([[1],[0],[0],[0]],dtype="int")
```

```python
#Menampilkan data x
x
```

```python
#Menampilkan Y
y
```

```python
# Melatih Model (fitting /Training )
history = model.fit (x, y,epochs= 500,verbose=2)
```

```python
# Prediksi Model
model.predict(x)
```

```python
#Membuat Hasil Pemodelan
model.predict(x).round()
```

```python
#Menampilkan nilai bobot
model.get_weights()
```
